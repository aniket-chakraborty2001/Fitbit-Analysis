```python
# Importing all required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
```

```python
# Explanatory Data Analysis (EDA) steps
# Reading the CSV data using Pandas library in a variable called data
data=pd.read_csv("/content/drive/MyDrive/Project_Data_Sets/FitBit Data.csv")


# Printing the type of columns used
print(data.dtypes)

# Printing the column names
print(data.columns)

# Printing the structure of the data frame
print(data.shape)

# Observig first 5 rows of the dataframe
data.head(5)
```

```python
# Data Manipulation Using Pandas Library
# Create a column called "Day" that contains 50 integers 1 to 50
data['Day'] = range(1,51)
data.head(5)

# To check if there are any missing values
print(data.isnull().sum().sum())

# Renaming the Column Day to Person
data.rename(columns = {'Day':'Person'},inplace = True)
data.columns
```

```python
# Distribution on Number of Sleeping Hours in the population Uisng Boxplot
plt.figure(figsize = (8,6))
sns.boxplot(x='Sleeping',data=data, color='gold', width =0.5)
plt.title('Distribution of Sleeping Hours in The Population')
plt.xlabel('Sleeping Hours',fontsize = 12)
plt.ylabel('Frequency',fontsize = 12)
plt.show()
```

```python
# Distribution of heartRate in the Population Uisng Boxplot
plt.figure(figsize = (8,6))
sns.boxplot(x='HeartRate',data=data, color='green', width =0.5)
plt.title('Distribution of HeartRate in The Population')
plt.xlabel('HeartRate',fontsize = 12)
plt.ylabel('Frequency',fontsize = 12)
plt.show()
```

```python
# Distribution of Blood Sugar Level in the Population Uisng Boxplot
plt.figure(figsize = (8,6))
sns.boxplot(x='Glucose',data=data, color='red', width =0.5)
plt.title('Distribution of Blood Sugar in The Population')
plt.xlabel('Blood Sugar',fontsize = 12)
plt.ylabel('Frequency',fontsize = 12)
plt.show()
```

```python
# Distribution of Cholesterol Level in the Population Uisng Boxplot
plt.figure(figsize = (8,6))
sns.boxplot(x='Cholesterol',data=data, color='aqua', width =0.6)
plt.title('Distribution of Fat in The Population')
plt.xlabel('Cholesterol',fontsize = 12)
plt.ylabel('Frequency',fontsize = 12)
plt.show()
```

```python
# Distribution of Weight in the Population Uisng Boxplot
plt.figure(figsize = (8,6))
sns.boxplot(x='Weight',data=data, color='orange', width =0.5)
plt.title('Distribution of Weight in The Population')
plt.xlabel('Weight',fontsize = 12)
plt.ylabel('Frequency',fontsize = 12)
plt.show()
```

```python
# Distribution of Height in the Population Uisng Boxplot
plt.figure(figsize = (8,6))
sns.boxplot(x='Height',data=data, color='skyblue', width =0.5)
plt.title('Distribution of Height in The Population')
plt.xlabel('Height',fontsize = 12)
plt.ylabel('Frequency',fontsize = 12)
plt.show()
```

```python
# Relationship Between HeartRate and Sleeping Hours Uisng Scatter plot
sleeping = data['Sleeping']
heartrate = data['HeartRate']
plt.plot(sleeping,heartrate,'ro')
plt.title('Relationship Between Sleeping and HeartRate')
plt.xlabel('Sleeping Hours',fontsize = 12)
plt.ylabel('Heart Rate',fontsize = 12)
plt.show()
```

```python
# Relationship Between Sleeping Hours,Sugar Level, and Cholesterol
sleeping = data['Sleeping']
sugar = data['Glucose']
fat = data['Cholesterol']
plt.plot(sleeping,sugar,color = 'red',linestyle = '-')
plt.plot(sleeping,fat,'g*')
plt.title('Interrealtion Between Slepping, Glucose and Fat')
plt.xlabel('Sleeping Hours',fontsize = 12)
plt.ylabel('Measurements',fontsize = 12)
plt.legend(['Glucose','Cholesterol'])
plt.show()
```

```python
# Relationship Between Disrance Run and Weight Using Scatter plot
run = data['Run']
weight = data['Weight']
plt.plot(run,weight,'bo')
plt.xlabel('Distance Run',fontsize = 12)
plt.ylabel('Weight of Person',fontsize = 12)
plt.title('Relationship Between Distance Run and Weight')
plt.show()
```

```python
# Change of BMI over Days Using Line Graph
bmi = data['BMI']
person = data['Person']
plt.plot(person,bmi,color='red',linestyle = '-')
plt.xlabel('The person Number')
plt.ylabel('The Body Mass Index')
plt.title('BMI over Number of Persons')
plt.show()
```

```python
# Track Changes in Weight of Persons using Line Graph
weight = data['Weight']
person = data['Person']
plt.plot(person, weight, color ='red',linestyle = '-')
plt.xlabel('Number of Persons',fontsize =12 )
plt.ylabel('Weight of Persons',fontsize = 12)
plt.title('Trend in Weight in The Population')
plt.show()
```

```python
# Track Changes in Height of Persons using Line Graph
height = data['Height']
person = data['Person']
plt.plot(person, height, color ='red',linestyle = '-')
plt.xlabel('Number of Persons',fontsize =12 )
plt.ylabel('Height of Persons (in Meters)',fontsize = 12)
plt.title('Trend of Height in The Population')
plt.show()
```

```python
# Represent the Distance Run with the Number of Persons using Bar Graph
run = data['Run']
person = data['Person']
plt.bar(person, run, color = 'green', width = 0.6)
plt.xlabel('Number of Person',fontsize = 12)
plt.ylabel('Distance Covered',fontsize = 12)
plt.title('Bar Graph of Distance Covered and Number of Persons')
plt.grid(color = 'black',linestyle = '--',alpha = 0.5, axis = 'y')
plt.figure(figsize = (10,12))
plt.show()
```

```python
# Creating Histogram to Know the Distribution of Sleeping Hours
sleeping = data['Sleeping']
plt.hist(sleeping , bins = 10, alpha = 0.8, color = 'blue', linewidth = 3)
style.use('fivethirtyeight')
plt.show()
```
